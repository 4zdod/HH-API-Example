```python
import requests
import pandas as pd

area = "40" # Initialize our area of searching

data = []  # Initialize list to store data

page = 0  # Initialize page number

while True:
    url = 'https://api.hh.ru/vacancies'
    params = {'area': area, 'page': page}
    response = requests.get(url, params=params)
    page_data = response.json()
    
    # Check if the 'items' key exists in the current page_data
    if 'items' in page_data:
        data.extend(page_data['items'])
        page += 1
    else:
        break

# Create a dataframe from data
df = pd.DataFrame(data)

# Save it to CSV file
csv_name = "all_jobs.csv"
df.to_csv(csv_name)

df = pd.read_csv("all_jobs.csv")

df
```
